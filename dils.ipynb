```python
#GRAFIK 
import pandas as pd
import matplotlib.pyplot as plt

# Ambil data dari GitHub
df = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/main/penjualan.csv")

# Hitung total per bulan
total_jan = df["jan"].sum()
total_feb = df["feb"].sum()
total_mar = df["mar"].sum()

# Buat dataframe untuk plotting
data_penjualan = pd.DataFrame({
    "Bulan": ["Januari", "Februari", "Maret"],
    "Total Penjualan": [total_jan, total_feb, total_mar]
})

# Plot
plt.figure(figsize=(10,6), facecolor="#f7f7f7")  
bars = plt.bar(
    data_penjualan["Bulan"], 
    data_penjualan["Total Penjualan"], 
    color=["#FFAEF8", "#C1ECA8", "#8ACCF8"], 
    alpha=0.9,
    edgecolor="black",
    linewidth=1.2
)

# Menambahkan judul
plt.title("TOTAL PENJUALAN PER BULAN (Januari  Maret)", fontsize=18, weight="bold", color="#333")

# Label sumbu
plt.xlabel("Bulan", fontsize=14, weight="bold")
plt.ylabel("Total Penjualan", fontsize=14, weight="bold")

# Tambahkan grid halus
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Tambahkan angka di atas batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f"{yval:,}", 
             ha="center", va="bottom", fontsize=12, weight="bold", color="#333")

plt.tight_layout()
plt.show()
```

Memvisualisasi Filter

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

#data
df = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/main/penjualan.csv")

#contoh filter: pilih nama yang mengandung "on" (case-insensitive) atau jan > 200
filtered = df[(df['nama'].str.contains('on', flags=re.IGNORECASE)) | (df['jan'] > 200)]

print("Data hasil filter:")
print(filtered)

# Plot hasil filter (misalnya Jan, Feb, Mar, Apr dibandingkan)
filtered.set_index("nama")[["jan", "feb", "mar", "apr"]].plot(
    kind="bar", figsize=(10,6), edgecolor="black"
)

plt.title("Data Terfilter (Nama mengandung 'on' atau Jan > 200)", fontsize=14, weight="bold")
plt.xlabel("Nama")
plt.ylabel("Jumlah Penjualan")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#tambahkan kolom kategori berdasarkan 'jan'
df['kategori_feb'] = df['feb'].apply(lambda x: "Tinggi" if x >= 250 else "Rendah")

#hitung jumlah masing-masing kategori
kategori_count = df['kategori_feb'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(kategori_count, labels=kategori_count.index, autopct='%1.1f%%', startangle=90, colors=["#CBFFBE","#FFA0A0"])
plt.title("Proporsi Kategori Penjualan Februari", fontsize=14, weight="bold")
plt.show()
```

```python
#scatter plot mar vs apr (hanya data terfilter)
plt.figure(figsize=(8,6))
plt.scatter(filtered['mar'], filtered['apr'], s=100, c="red", alpha=0.7, edgecolors="black")

plt.title("Scatter Plot Penjualan Mar vs Apr (Filtered)", fontsize=14, weight="bold")
plt.xlabel("Penjualan Maret")
plt.ylabel("Penjualan April")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
